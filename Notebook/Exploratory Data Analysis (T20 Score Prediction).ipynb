```python
# Import the important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
import pickle
```

```python
# Load the dataset
df = pickle.load(open('data_level3.pkl','rb'))
```

```python
# Call the dataframe
df
```

```python
# Countplot
for i in df.columns:
    plt.figure(figsize=(15,6))
    sns.countplot(x=df[i], data = df, palette = 'hls')
    plt.xticks(rotation = 90)
    plt.show()
```

```python
# To get the insights about the distribution of each features
for i in df.columns:
    plt.figure(figsize=(15,6))
    sns.histplot(df[i], kde = True, palette = 'hls')
    plt.xticks(rotation = 90)
    plt.show()
```

```python
df_new = df[["batting_team","bowling_team","city"]]
```

```python
for i in df_new:
    plt.figure(figsize=(30,20)) 
    plt.pie(df[i].value_counts(), labels=df[i].value_counts().index, autopct='%1.1f%%', textprops={ 'fontsize': 25,
                                                                                                           'color': 'black',
                                                                                                           'weight': 'bold',
                                                                                                           'family': 'serif' }) 
    hfont = {'fontname':'serif', 'weight': 'bold'}
    plt.title(i, size=20, **hfont) 
    plt.show()
```

```python
cor_matrix = df.corr()
```

```python
plt.style.use("default")
plt.figure(figsize=(10, 8))
# setting the parameter values
cmap = "tab20"
center = 0
# setting the parameter values
linewidths = 2
linecolor = "red"
# Plotting the heatmap
sns.heatmap(cor_matrix,cmap=cmap, center=center,linewidths=linewidths,linecolor=linecolor,annot=True)
plt.show()
```
